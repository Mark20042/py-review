## Dictionary

keys are unique !

don't have order

uses curly brackets {} and colon :

```python
my_dict = { 1:2 , "aaluhh" : 4 , 2:3}

print(type(my_dict))

print(my_dict[1]) ## the format is the key : value

## we cant get the key when not declared like my_dict[2]


print(1 in my_dict) # to check if that key exist will throw a boolean



# can also use the get() function to find that value
print(my_dict.get("aaluhh"))
```

Dictionary Functions

```python
## Clear

print(my_dict)
my_dict.clear() ## will clear the dictionary
print(my_dict)

## Copy
grocery_dict = { 'Melon' : [5,9], 'Apple ': 3 , "Strawberry ": 5 , "Orange": 9}
print(grocery_dict)
grocery_dict_copy = grocery_dict.copy() ## will get a copy of a dictionary
print(grocery_dict_copy)

## From key function
list_keys = ['Apple', 'Strawberry']
new_dict = dict.fromkeys(list_keys, 20) ## to replace the value in the dict
print(new_dict)

## items Function
print(grocery_dict.items()) # will return all the dictionaries in tuple format


## Keys Function
print(grocery_dict.keys()) # will return all the keys in tuple format
```

Looping the dictionaries

```python
for k, v in grocery_dict.items():
  print(k,v)
```

## Set

use curly brackets {}

elements can only be unique

cant be indexed

dont have order

mutable

removes duplicate element

will order in ascending



```python
my_set = { 9,5,7,4,2,1}
print(type(my_set))
my_set.add(8)
# my_set.pop() # will remove first element from the set
my_set.remove(4) # will remove element from the set raise error when element not found
my_set.discard(5) # will remove element from the set does not raise a error when not element found
print(my_set)
```
